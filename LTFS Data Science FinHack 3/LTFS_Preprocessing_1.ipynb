```python
def get_avg(col_name):
    div = col_name.str.split(',', expand = True).replace('',0)
    ch_typ = div.astype('float')
    avg = ch_typ.assign(average = ch_typ.mean(axis=1, numeric_only=True))
    return avg['average']

temp['AMT PAID - HIST'] = get_avg(temp['AMT PAID - HIST'])
```

```python
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('Weekly', 4.33)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('BiWeekly', 2.167)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('Monthly', 1)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('BiMonthly', 0.5)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('Quarterly', 0.25)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('Semi annually', 0.167)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('Annually', 0.0833)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('Single Payment Loan (bullet / balloon)', 1)
temp['INSTALLMENT-PERIOD'] = temp['INSTALLMENT-PERIOD'].replace('Other', 4.33)
```
